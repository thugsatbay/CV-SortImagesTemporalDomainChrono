```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open('store_test_1.pkl', 'rb') as f:
    plot = pickle.load(f)
plot = plot[0]
```

```python
for k,v in plot.items():
    plot[k] = v[::-1]
```

```python
n_groups = len(plot)
correct_bar = map(lambda x : x[1][0], plot.items())
wrong_bar = map(lambda x : x[1][1], plot.items())
 
# create plot
fig = plt.figure(figsize=(25, 15))
ax = plt.subplot(111)
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = ax.bar(index, correct_bar, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Correct')
 
rects2 = ax.bar(index + bar_width, wrong_bar, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Wrong')

labels_right = [str(int((float(v[0])/(v[0]+v[1]))*100)) + '%' for k,v in plot.items()]
#labels_wrong = [str(int((float(v[1])/(v[0]+v[1]))*100)) + '%' for k,v in plot.items()]

rects = ax.patches
    
for rect, label in zip(rects, labels_right):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label, ha='center', va='bottom')

print "Plotted graphs for both correct and wrong", float(sum([int((float(v[0])/(v[0]+v[1]))*100) for k,v in plot.items()])) / len(plot)
plt.xlabel('Class')
plt.ylabel('Samples')
plt.title('Correct Wrong - 1')
plt.xticks(index + bar_width, tuple(plot.keys()), rotation=90)
plt.legend()
plt.savefig('abc.jpg')
```
